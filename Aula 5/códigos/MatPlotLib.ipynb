```python
# Box e Whisker Plot
# Boxplot  um mtodo grfico para representar a variao de uma varivel ao longo de um conjunto de dados.
# O boxplot  formado pelo primeiro e terceiro quartis (limites da caixa), 
# pela mediana (linha no interior da caixa) 
# e pelos limites inferior e superior (linhas verticais que se estendem a partir da caixa).
uniformSkewed = np.random.rand(100) * 100 - 40
high_outliers = np.random.rand(10) * 50 + 100
low_outliers = np.random.rand(10) * -50 - 100
data = np.concatenate((uniformSkewed, high_outliers, low_outliers))
plt.boxplot(data)
plt.show()
```
