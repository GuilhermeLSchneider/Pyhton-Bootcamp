```python
from matplotlib.pyplot import scatter
import numpy as np
from statistics import mean
from numpy import dot

# O numpy possui uma função para calcular a covariância, 
# mas vamos implementar a nossa própria função para entendermos melhor o conceito
def de_mean(x):
    xmean = mean(x)
    return [xi - xmean for xi in x]

def covariance(x, y):
    n = len(x)
    return dot(de_mean(x), de_mean(y)) / (n - 1)

np.random.seed(50)
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000)

scatter(pageSpeeds, purchaseAmount)

covariance(pageSpeeds, purchaseAmount)
```

```python
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

scatter(pageSpeeds, purchaseAmount)

covariance(pageSpeeds, purchaseAmount)
# Desse jeito, a covariância não é muito útil, sendo um pouco complicado de entender o que esse valor significa.
# Para isso, iremos utilizar a correlação, que é um valor entre -1 e 1, que indica a relação entre as duas variáveis.
```

```python
def correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x, y) / stddevx / stddevy

correlation(pageSpeeds, purchaseAmount)
```

```python
np.corrcoef(pageSpeeds, purchaseAmount)
# Para ler o resultado obtido, basta olhar a matriz retornada, 
# onde a diagonal principal é a correlação de cada variável com ela mesma.
```

```python
purchaseAmount = 100 - pageSpeeds * 3

scatter(pageSpeeds, purchaseAmount)

correlation(pageSpeeds, purchaseAmount)
```

```python
np.cov(pageSpeeds, purchaseAmount), # Calcula a matriz de covariância, que não é muito útil para entender a relação entre as variáveis
np.corrcoef(pageSpeeds, purchaseAmount) # Calcula a matriz de correlação, que é mais útil para entender a relação entre as variáveis
# onde, nesse caso, indica que a correlação entre as variáveis é -1, ou seja, são inversamente proporcionais.
```
