```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(150)
vals = np.random.normal(0, 0.5, 10000)

plt.hist(vals, 50)
plt.show()
```

```python
np.percentile(vals, 99)
```

```python
np.percentile(vals, 50) # Um pouco deslocado para a esquerda do zero
```
