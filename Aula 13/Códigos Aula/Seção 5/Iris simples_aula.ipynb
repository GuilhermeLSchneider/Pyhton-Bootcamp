# Deep Learning A Z Classificação Mais Classes - Iris simples

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import sklearn
```

```python
pd.__version__, np.__version__, tf.__version__, sklearn.__version__
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import utils as np_utils
from sklearn.preprocessing import LabelEncoder # Será usado para transformar as labels em números
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
base = pd.read_csv('iris.csv')
base
```

```python
X = base.iloc[:, 0:4].values # Atributos previsores, 150 linhas e 4 colunas
X
```

```python
y = base.iloc[:, 4].values # Atributos classe, 150 linhas e 1 coluna
y
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y) # Transforma as labels em números
y
```

```python
y.shape
```

```python
y = np_utils.to_categorical(y) # Transforma os valores de y em categorias 
# (1, 0, 0 indica a classe da Iris-setosa, 0, 1, 0 indica a classe da Iris-versicolor e 0, 0, 1 indica a classe da Iris-virginica)
y
```

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size=0.25)
```

```python
X_treinamento.shape, X_teste.shape
```

```python
y_treinamento.shape, y_teste.shape
```

```python
(4 + 3) / 2
# (Numero de atributos (features) + Numero de classes (resultado)) / 2 = 3.5
```

```python
rede_neural = Sequential([
    tf.keras.layers.InputLayer(shape = (4,)), # Camada de entrada com 4 neurônios (4 atributos)
    tf.keras.layers.Dense(units = 4, activation = 'relu'),
    tf.keras.layers.Dense(units = 4, activation = 'relu'),
    tf.keras.layers.Dense(units = 3, activation = 'softmax') # Note que aqui utilzamos a função de ativação softmax, por conta de ser um problema de classificação com mais de duas classes
])
```

```python
rede_neural.summary()
```

```python
rede_neural.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])
# Função de perda e métricas para problemas de classificação com mais de duas classes
```

```python
rede_neural.fit(X_treinamento, y_treinamento, batch_size = 10, epochs = 1000)
```

```python
rede_neural.evaluate(X_teste, y_teste) # Avalia o modelo com os dados de teste
```

```python
previsoes = rede_neural.predict(X_teste) # Faz as previsões com os dados de teste
print(previsoes)
```

```python
previsoes = previsoes > 0.5
print(previsoes)
```

```python
y_teste2 = [np.argmax(t) for t in y_teste] # Retorna o índice do valor True em cada linha
print(y_teste2)
```

```python
previsoes2 = [np.argmax(t) for t in previsoes]
print(previsoes2)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_teste2, previsoes2)
```
