```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('image.png')
```

```python
plt.imshow(img)
```

```python
rgb_image = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

```python
gray_image = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

```python
img_gray_small = cv.resize(gray_image, (50,50)) # redimensiona a imagem pela metade (50% em ambos os eixos)
plt.imshow(img_gray_small, cmap='gray')
```

```python
plt.imshow(gray_image[10:340, :], cmap='gray') # recorta a imagem, mantendo apenas a parte superior
```

```python
plt.imshow(gray_image[:, :300], cmap='gray') # recorta a imagem, mantendo apenas ate a coluna 300
```
