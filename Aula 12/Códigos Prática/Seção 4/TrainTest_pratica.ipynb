```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import scatter

np.random.seed(2)

pageSpeeds = np.random.normal(3.0, 1.0, 100)
purchaseAmount = np.random.normal(50.0, 30.0, 100) / pageSpeeds

scatter(pageSpeeds, purchaseAmount)
```

```python
# Dividindo os dados em treino e teste (80% treino, 20% teste)
# Importante salientar que a divisão não deve ser feita dessa forma,
# mas sim utilizando funções de shuffle, para evitar que os dados de treino e teste sejam muito parecidos
trainX = pageSpeeds[:80]
testX = pageSpeeds[80:]

trainY = purchaseAmount[:80]
testY = purchaseAmount[80:]

scatter(trainX, trainY)
```

```python
scatter(testX, testY)
```

```python
x = np.array(trainX)
y = np.array(trainY)

p4 = np.poly1d(np.polyfit(x, y, 6)) # Utilizando agora um polinômio de grau 6
```

```python
xp = np.linspace(0, 7, 100)
axes = plt.axes()
axes.set_xlim([0, 7])
axes.set_ylim([0, 200])
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
testx = np.array(testX)
testy = np.array(testY)

axes = plt.axes()
axes.set_xlim([0, 7])
axes.set_ylim([0, 200])
plt.scatter(testx, testy)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(testy, p4(testx))

print(r2)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(np.array(trainY), p4(np.array(trainX)))

print(r2)
```

* Aparentemente um polinômio de grau 5 ou 6 parece ser a melhor opção para esse caso
