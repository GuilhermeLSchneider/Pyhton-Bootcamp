```python
import numpy as np
from numpy import array

def createClusteredData(N, k):
    np.random.seed(10)
    pointsPerCluster = float(N)/k # Define os números de pontos por cluster
    X = []
    # Cria os clusters
    for i in range(k):
        incomeCentroid = np.random.uniform(20000.0, 200000.0) # Define o centro de renda entre 20k e 200k
        ageCentroid = np.random.uniform(20.0, 70.0) # Define o centro de idade entre 20 e 70 anos
        for j in range(int(pointsPerCluster)): # Cria os pontos
            X.append([np.random.normal(incomeCentroid, 10000.0), np.random.normal(ageCentroid, 2.0)]) # Cria os pontos com base no centro e desvio padrão
    X = array(X)
    return X
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale

data = createClusteredData(100, 5)

model = KMeans(n_clusters=5)

model = model.fit(scale(data)) # Normaliza os dados (importante para encontrar bons resultados)

print(model.labels_)

plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], # Renda
            data[:,1], # Idade
            c=model.labels_.astype(float))
plt.show()
```

```python
# Com o resultado acima, parece que utilizando 4 clusters irá funcionar também 
# (no entanto, não sabemos as características dos clusters, pode ser que o agrupamento seja incorreto com 4 clusters)
model = KMeans(n_clusters=4)

model = model.fit(scale(data)) # Normaliza os dados (importante para encontrar bons resultados)

print(model.labels_)

plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], # Renda
            data[:,1], # Idade
            c=model.labels_.astype(float))
plt.show()
```
