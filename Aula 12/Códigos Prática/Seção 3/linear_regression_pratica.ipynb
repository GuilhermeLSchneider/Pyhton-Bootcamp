```python
import numpy as np
from matplotlib.pyplot import scatter
```

```python
np.random.seed(2) # Seed para garantir que os valores aleatórios sejam sempre os mesmos

pageSpeeds = np.random.normal(3.0, 1.0, 1000) # Distribuição normal centrada no 3.0 com desvio padrão de 1.0 com 1000 valores
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.5, 1000)) * 3 # Aumentamos a variância do ruído para 0.5

scatter(pageSpeeds, purchaseAmount) # Gráfico de dispersão
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount) # Regressão linear
```

```python
r_value ** 2 # Coeficiente de determinação
# Nesse caso, o coeficiente resultou em 0.8077, o que indica que 80.77% da variação de purchaseAmount é explicada por pageSpeeds
# Ou seja, nao é uma relação perfeita, mas é uma relação forte
```

```python
import matplotlib.pyplot as plt

# Função para prever valores
def predict(x):
    return slope * x + intercept # y = mx + b

fitLine = predict(pageSpeeds)

# Gráfico de dispersão com a linha de regressão
plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitLine, c='r')
plt.show()
```
