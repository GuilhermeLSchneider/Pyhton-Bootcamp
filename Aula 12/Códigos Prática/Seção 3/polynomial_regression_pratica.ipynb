```python
import numpy as np
from matplotlib.pyplot import scatter 
import matplotlib.pyplot as plt
```

```python
np.random.seed(2)
pageSpeeds = np.random.normal(3.0, 1.0, 1000) # 1000 números aleatórios com média 3 e desvio padrão 1
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds # O purchaseAmount é inversamente proporcional ao pageSpeeds

scatter(pageSpeeds, purchaseAmount)
```

```python
# Converte as listas em um array
x = np.array(pageSpeeds)
y = np.array(purchaseAmount)

p3 = np.poly1d(np.polyfit(x, y, 3)) # Ajusta um polinômio de grau 3
p4 = np.poly1d(np.polyfit(x, y, 4)) # Ajusta um polinômio de grau 4
p5 = np.poly1d(np.polyfit(x, y, 5)) # Ajusta um polinômio de grau 5
p6 = np.poly1d(np.polyfit(x, y, 6)) # Ajusta um polinômio de grau 6
```

```python
xp = np.linspace(0, 7, 100) # Gera 100 pontos entre 0 e 7 para plotar o gráfico
plt.scatter(x, y)
plt.plot(xp, p3(xp), c='g')
plt.show()

plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()

plt.scatter(x, y)
plt.plot(xp, p5(xp), c='y')
plt.show()

plt.scatter(x, y)
plt.plot(xp, p6(xp), c='m')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2_p3 = r2_score(y, p3(x))
r2_p4 = r2_score(y, p4(x))
r2_p5 = r2_score(y, p5(x))
r2_p6 = r2_score(y, p6(x))
```

```python
r2_p3, r2_p4, r2_p5, r2_p6
```

```python
# O polinômio de grau 3 acaba não guardando a variância dos dados
# O polinômio de grau 4 é o que melhor se ajusta aos dados
# O polinomio de grau 5 e 6 acabam overfitando os dados
```
