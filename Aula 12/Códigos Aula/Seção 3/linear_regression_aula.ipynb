```python
import numpy as np
from matplotlib.pyplot import scatter
import matplotlib.pyplot as plt
```

```python
np.random.seed(11) # Seed para garantir que os valores aleatórios sejam sempre os mesmos

pageSpeeds = np.random.normal(3.0, 1.0, 1000) # Distribuição normal centrada no 3.0 com desvio padrão de 1.0 com 1000 valores
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3 # Essa fórmula é para simular uma relação negativa entre pageSpeeds e purchaseAmount

scatter(pageSpeeds, purchaseAmount) # Gráfico de dispersão
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount) # Regressão linear
```

```python
r_value ** 2 # Coeficiente de determinação
# Apresentou valor próximo a 1, o que indica que a regressão linear é um bom modelo para esses dados
```

```python
import matplotlib.pyplot as plt

# Função para prever valores
def predict(x):
    return slope * x + intercept # y = mx + b

fitLine = predict(pageSpeeds)

# Gráfico de dispersão com a linha de regressão
plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitLine, c='r')
plt.show()
```

```python
np.random.seed(3) # Seed para garantir que os valores aleatórios sejam sempre os mesmos

# Criando um conjunto outro cunjunto de dados sobre veículos e consumo de combustível
horsepower = np.random.normal(150.0, 20.0, 500) # Distribuição normal centrada no 150.0 com desvio padrão de 20.0 com 100 valores
mpg = 150 - horsepower*0.7 + np.random.normal(0, 4, 500) # Essa fórmula é para simular uma relação negativa entre horsepower e mpg

plt.xlabel('Horsepower (HP)')
plt.ylabel('Miles per Galon (MPG)')
plt.title('Horsepower vs MPG')
scatter(horsepower, mpg) # Gráfico de dispersão
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(horsepower, mpg) # Regressão linear

# Função para prever valores
def predict(x):
    return slope * x + intercept # y = mx + b

fitLine = predict(horsepower)

# Gráfico de dispersão com a linha de regressão
plt.scatter(horsepower, mpg)
plt.xlabel('Horsepower (HP)')
plt.ylabel('Miles per Galon (MPG)')
plt.title('Horsepower vs MPG')
plt.plot(horsepower, fitLine, c='r')
plt.show()
```
