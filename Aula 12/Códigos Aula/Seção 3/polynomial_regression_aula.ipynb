```python
import numpy as np
from matplotlib.pyplot import scatter 
import matplotlib.pyplot as plt
```

```python
np.random.seed(2)
pageSpeeds = np.random.normal(3.0, 1.0, 1000) # 1000 números aleatórios com média 3 e desvio padrão 1
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds # O purchaseAmount é inversamente proporcional ao pageSpeeds

scatter(pageSpeeds, purchaseAmount)
```

```python
# Converte as listas em um array
x = np.array(pageSpeeds)
y = np.array(purchaseAmount)

p4 = np.poly1d(np.polyfit(x, y, 4)) # Ajusta um polinômio de grau 4
```

```python
xp = np.linspace(0, 7, 100) # Gera 100 pontos entre 0 e 7 para plotar o gráfico
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
# Aqui é importante analisar o gráfico para ver se o polinômio de grau 4 é o mais adequado para o ajuste
# se ele estiver tentando se ajustar a pontos que não fazem sentido, é melhor tentar um polinômio de grau menor
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, p4(x))

print(r2) # Quanto mais próximo de 1, melhor o ajuste (claro, cuidando para não overfitar)
```

```python
np.random.seed(5) # Seed para garantir que os valores aleatórios sejam sempre os mesmos

# Criando um conjunto outro cunjunto de dados sobre veículos e consumo de combustível com regressão polinomial
horsepower = np.random.normal(200.0, 30.0, 500) # Distribuição normal centrada no 150.0 com desvio padrão de 20.0 com 100 valores
mpg = np.random.normal(50.0, 1.0, 500) * 100 / horsepower # Consumo de combustível inversamente proporcional à potência

plt.xlabel('Horsepower (HP)')
plt.ylabel('Miles per Galon (MPG)')
plt.title('Horsepower vs MPG')
scatter(horsepower, mpg) # Gráfico de dispersão
```

```python
x = np.array(horsepower)
y = np.array(mpg)

p4 = np.poly1d(np.polyfit(x, y, 3)) # Ajuste polinomial de grau 4
```

```python
xp = np.linspace(95, 300, 100) # Gera 100 pontos entre 0 e 7 para plotar o gráfico
plt.scatter(x, y)
plt.xlabel('Horsepower (HP)')
plt.ylabel('Miles per Galon (MPG)')
plt.title('Horsepower vs MPG')
plt.plot(xp, p4(xp), c='r')
plt.show()
# Aqui é importante analisar o gráfico para ver se o polinômio de grau 4 é o mais adequado para o ajuste
# se ele estiver tentando se ajustar a pontos que não fazem sentido, é melhor tentar um polinômio de grau menor
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(horsepower, mpg) # Regressão linear

# Função para prever valores
def predict(x):
    return slope * x + intercept # y = mx + b

fitLine = predict(horsepower)
# Gráfico de dispersão com a linha de regressão
plt.scatter(horsepower, mpg)
plt.xlabel('Horsepower (HP)')
plt.ylabel('Miles per Galon (MPG)')
plt.title('Horsepower vs MPG')
plt.plot(horsepower, fitLine, c='r')
plt.show()
```
