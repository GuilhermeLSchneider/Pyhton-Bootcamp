```python
import numpy as np
from numpy import array

def createClusteredData(N, k):
    np.random.seed(10)
    pointsPerCluster = float(N)/k # Define os números de pontos por cluster
    X = []
    # Cria os clusters
    for i in range(k):
        incomeCentroid = np.random.uniform(20000.0, 200000.0) # Define o centro de renda entre 20k e 200k
        ageCentroid = np.random.uniform(20.0, 70.0) # Define o centro de idade entre 20 e 70 anos
        for j in range(int(pointsPerCluster)): # Cria os pontos
            X.append([np.random.normal(incomeCentroid, 10000.0), np.random.normal(ageCentroid, 2.0)]) # Cria os pontos com base no centro e desvio padrão
    X = array(X)
    return X
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale

data = createClusteredData(100, 5)

model = KMeans(n_clusters=5)

model = model.fit(scale(data)) # Normaliza os dados (importante para encontrar bons resultados)

print(model.labels_)

plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], # Renda
            data[:,1], # Idade
            c=model.labels_.astype(float))
plt.show()
```

```python
import numpy as np
from numpy import array

def createClusteredData2(N, k):
    np.random.seed(3)
    pointsPerCluster = float(N)/k # Define os números de pontos por cluster
    X = []
    # Cria os clusters
    for i in range(k):
        transactions = np.random.uniform(-100000, 100000)
        productsBought = np.random.uniform(0.0, 70.0)
        for j in range(int(pointsPerCluster)): # Cria os pontos
            X.append([np.random.normal(transactions, 5000.0), np.random.normal(productsBought, 2.0)]) # Cria os pontos com base no centro e desvio padrão
    X = array(X)
    return X
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale

data = createClusteredData2(200, 3)

model = KMeans(n_clusters=4)

model = model.fit(scale(data)) # Normaliza os dados (importante para encontrar bons resultados)

print(model.labels_)

plt.figure(figsize=(8, 6))
plt.scatter(data[:,0], # Renda
            data[:,1], # Idade
            c=model.labels_.astype(float))
plt.xlabel('Transações executadas (Valores negativos indicam devoluções)')
plt.ylabel('Número de produtos comprados (em centenas)')
plt.show()
```
