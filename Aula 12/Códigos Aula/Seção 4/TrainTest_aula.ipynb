```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import scatter

np.random.seed(2)

pageSpeeds = np.random.normal(3.0, 1.0, 100)
purchaseAmount = np.random.normal(50.0, 30.0, 100) / pageSpeeds

scatter(pageSpeeds, purchaseAmount)
```

```python
horsepower = np.random.normal(150.0, 20.0, 500) # Distribuição normal centrada no 150.0 com desvio padrão de 20.0 com 100 valores
mpg = 150 - horsepower*0.7 + np.random.normal(0, 4, 500) # Essa fórmula é para simular uma relação negativa entre horsepower e mpg
scatter(horsepower, mpg)
```

```python
trainX = horsepower[:400]
testX = horsepower[400:]

trainY = mpg[:400]
testY = mpg[400:]

scatter(trainX, trainY)
scatter(testX, testY)
```

```python
# Dividindo os dados em treino e teste (80% treino, 20% teste)
# Importante salientar que a divisão não deve ser feita dessa forma,
# mas sim utilizando funções de shuffle, para evitar que os dados de treino e teste sejam muito parecidos
trainX = pageSpeeds[:80]
testX = pageSpeeds[80:]

trainY = purchaseAmount[:80]
testY = purchaseAmount[80:]

scatter(trainX, trainY)
scatter(testX, testY)
```

```python
scatter(testX, testY)
```

```python
x = np.array(trainX)
y = np.array(trainY)

p4 = np.poly1d(np.polyfit(x, y, 8)) # Utilizando um polinômio de grau 8 (causará overfitting)
```

```python
xp = np.linspace(0, 7, 100)
axes = plt.axes()
axes.set_xlim([0, 7])
axes.set_ylim([0, 200])
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
testx = np.array(testX)
testy = np.array(testY)

axes = plt.axes()
axes.set_xlim([0, 7])
axes.set_ylim([0, 200])
plt.scatter(testx, testy)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(testy, p4(testx))

print(r2) # O valor de r2 é muito baixo, indicando que o modelo está sofrendo de overfitting
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(np.array(trainY), p4(np.array(trainX)))

print(r2) # Aqui já é possível verificar que o valor de r2 não é tão alto já nos dados de treino
```
